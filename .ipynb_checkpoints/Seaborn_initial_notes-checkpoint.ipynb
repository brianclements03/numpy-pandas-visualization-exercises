```python
#scatterplot in matplotlib
plt.scatter(tips.total_bill, tips.tip)
#total bill on x, tip on y
```

```python
#explore this relationship using relplot
sns.relplot(data = tips, x = 'total_bill', y = 'tip')
#notice the labels
```

```python
#update defaults (rc params in matplotlib)
sns.set() #alias for sns.set_themes
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip')

```

hue, size and style arguments

```python
#visualize if relationship is different for smoker v non (use hue arg)

sns.relplot(data = tips, x = 'total_bill', y = 'tip', 
            hue = 'smoker',
           style = 'sex',
           size = 'size')
#holy fucking shit
```

### Relplot with 'kind' argument

```python
#lineplot for total bill vs tips - may not be best in this cae.  note use of 'kind' arg

sns.relplot(data=tips, x = 'total_bill', y= 'tip', kind = 'line')
```

```python
#lineplot for discrete values.  the lineplot will use mean value at each size to draw a line
#shaded region is 95% Conf Interval. Can be turned off w 'ci = None'

sns.relplot(data = tips, x = 'size', y = 'tip', kind = 'line') #, estimator = 'mean'/'max'/'min', etc)
```

https://en.wikipedia.org/wiki/Bootstrapping.    (statistics) confidence intervals i guess

```python
#Facetgrid with col

sns.relplot(data = tips, x = 'total_bill', y = 'tip', 
            col = 'day',
           hue = 'time',
           #col_wrap = 2,
           col_order = ['Thur','Fri','Sat', 'Sun'])

#wow.
```

```python
#facetgrid w lineplot

sns.relplot(data = tips, x = 'size', y = 'tip', col = 'day', kind = 'line')

```

## Distributions: displot

```python
#displot for total_bill (his, kide, rug)

sns.displot(data = tips, x = 'tip', kde = True, rug = True)

```

```python
#histogram with hue and stacking and palette

sns.displot(data = tips, x = 'total_bill',
           hue = 'sex',
           multiple = 'stack',
            palette = ['b','r']
           )

```

```python
#use col arg with 'sex' to creat 'small multiples'

sns.displot(data = tips, x = 'total_bill', col = 'sex',
           color = 'b', alpha = .4)

```

```python
#histplot weill return an axes level obj

sns.histplot(data = tips, x = 'tip')
```

```python
# Set kind = 'kde'. kde = probability of finding a observartions at particular value of x. 
# Smooth out version of histogram. Estimating probabilty density function (PDF) in a non-parametric way.

sns.kdeplot(data = tips, x = 'total_bill',
           hue = 'sex',
           multiple = 'stack',
            palette = ['b', 'r']
           )
```

```python
#we can make boxplot with kind - 'box' arg.  figure level object (vs axes level)

sns.catplot(data = tips, y = 'tip', kind = 'box')
```

### Heatmap

```python
#crosstab of time vs smoker

ctab = pd.crosstab(tips.day, tips.smoker)
ctab
```

```python
#create the heatmap

sns.heatmap(ctab, cmap = 'Blues', annot = True)
```

sequential vs diverging colors

```python
#heatmap using mpg dataset

mpg = data('mpg')
mpg.head()
```

```python
#calculate correlation for num variables using pandas's .corr() method

mpg.corr()
```

```python
labels = ['Displacement', 'Model Year', 'cylinders', 'City MPG', 'Highway MPG']
```

```python
#heatmap for correlation table above

sns.heatmap(mpg.corr(), cmap = 'RdBu_r',   #vs 'RdBu'
           center = 0,
           vmax = 1,
           vmin = -1,
            annot = True,
            fmt = '.2f',
            linewidth = 1,
            linecolor = 'k',
            xticklabels = labels,
            yticklabels = labels
           )  
#eg of a divergent color scheme
```
