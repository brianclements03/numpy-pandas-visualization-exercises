## And now... SQL and Python/Pandas together at last
- You might need to install pymysql with `python -m pip install pymysql`
### Make sure to have a .gitignore and that it lists env.py
- Then we'll make our env.py file containing 3 plain old strings stored to variables:
    - `host = "ip.of.sql.server"`
    - `user = "your_username"`
    - `password = "your_password"`
- Then we'll import our values from env.py
- Creat a connection string 
- Create a SQL query like "select * from employees" in a plain old string
- Use `pd.read_sql(query, url)` to have pandas run the SQL and return a df

```python
#let's make a function that takes in a database name as a string
#this functions also performs our imports from env
def get_db_url(db_name):
    from env import user, host, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
```

```python
sql = """
    SELECT * from departments limit 3000
"""
```

```python
#called the "connection string" bc it has all the info to connect
url = get_db_url('employees')
#don't print this ^^
df = pd.read_sql(sql,url)
df.head()
```

### .merge

- Works like a SQL join

`left_df.merge(right_df, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, indicator=False)`

```python
#See curriculum for notes on what all the arguments mean. 
```

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
#notice the indicator argument, which provides useful info on where the column came from : )
```

```python
(users.merge(roles, 
            left_on='role_id', 
            right_on='id', 
            how='outer')
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id', 
                     'name_x': 'employee',
                     'id_y': 'role_id',
                     'name_y': 'role'}
            )
)
#if you want to drop those name_x/y columns
```

```python
n = 40

orders = pd.DataFrame({
    'drink': np.random.choice(['Tea', 'Water', 'Water'], n),
    'meal': np.random.choice(['Curry', 'Yakisoba Noodle', 'Pad Thai'], n),
})

orders.sample(10)
```

```python
prices = {
    'Yakisoba Noodle': 9,
    'Curry': 11,
    'Pad Thai': 10,
    'Tea': 2,
    'Water': 0,
}
```

```python
orders.drink.map(prices)
```

```python
orders.meal.map(prices)
```

```python
orders['bill'] = orders.meal.map(prices) + orders.drink.map(prices)
```

```python
orders.head()
```

```python
orders.pivot_table(index='meal', columns='drink', values='bill')
```
