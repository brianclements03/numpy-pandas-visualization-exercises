```python
#NEXT UP: LETS GET DIFFERENT N VALUES FOR DIFFERENT EQUATIONS SO THAT IT'S BETTER VISUALLY
#EG THE FACTORIAL ONLY TAKES INTEGERS; WAS THINKING OF USING NP.LINSPACE TO GET MANY MORE ITERATIONS FROM 1 TO 5


n = [1,2,3,4,5]
y1 = [(0 * n+ 1) for n in n]
y2 = [math.log(n) for n in n]
y3 = [n for n in n]
y4 = [(n * math.log(n)) for n in n]
y5 = [(n ** 2) for n in n]
y6 = [(2 ** n) for n in n]
y7 = [(math.factorial(n)) for n in n]
y8 = [(n ** n) for n in n]




plt.figure(figsize = (10,8))

plt.plot(n, y1,
        label = 'O(1)')
plt.plot(n, y2,
        label = 'O(log n)')
plt.plot(n, y3,
        label = 'O(n)')
plt.plot(n, y4,
        label = 'O(n log n)')
plt.plot(n, y5,
        label = 'O($n^2$)')
plt.plot(n, y6,
        label = 'O($2^n$)')
plt.plot(n, y7,
        label = 'O($n!$)')
plt.plot(n, y8,
        label = 'O($n^n$)')




plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.title('Big O Notation')
plt.legend() 
plt.show()
```
