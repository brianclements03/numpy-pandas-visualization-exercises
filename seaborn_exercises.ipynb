### Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

```python
swiss['is_catholic'] = np.where(swiss.Catholic >= 70, True, False)
swiss.head()
```

### Does whether or not a province is Catholic influence fertility?

```python
sns.scatterplot(data = swiss, x = 'Catholic', y = 'Fertility', hue = 'is_catholic')
```

```python
sns.barplot(data = swiss, x = 'is_catholic', y = 'Fertility')
```

### Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from get_db_url import get_db_url
url = get_db_url('chipotle')
sql = """
    SELECT * from orders
"""
chipotle_df = pd.read_sql(sql,url)
chipotle_df.head()
```

```python
#chipotle_df['item_price'] = chipotle_df.item_price.apply(lambda value : float(value[1:].strip()))
# i ran the above code to get those item_price strings into float format
chipotle_df.sample()
```

```python
most_popular = chipotle_df.pivot_table(aggfunc='sum', index='item_name').reset_index().sort_values(by='quantity', ascending = False).head(4)
most_popular
```

```python
sns.barplot(data = most_popular, x = 'item_name', y = 'item_price')
plt.xlabel("Item", )
plt.ylabel("Revenue")
plt.tight_layout()
```
