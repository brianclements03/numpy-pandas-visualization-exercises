Add an anotation for the point 0, 0, the origin.

```python
x = range(-100, 100)
y = [(n ** 2 - n + 2) for n in x]

plt.plot(x,y)
plt.title('$y=x^2-x+2$')
plt.annotate('This Point', 
             xy = (0, 0), 
             xytext = (1, 3000),
             arrowprops = dict(facecolor = 'gray'))
#plt.savefig('figure1.png');
```

```python
x1 = [x * 1 for x in range(0,10)]
y1 = [math.sqrt(x) for x in x1]

# 100 linearly spaced numbers
x2 = np.linspace(-5,5,10)
y2 = [x ** 3 for x in x2]
x3 = np.linspace(-5,5,100)
y3=[2**x for x in x3]
x4 = np.linspace(-10,10,100)
y4 = [1/(x+1) for x in x4]



plt.subplot(2,2,1)
plt.plot(x1,y1, 
         marker = 'v',
        label = '$\sqrt{x}$')
plt.title('$\sqrt{x}$')
plt.legend()

# setting the axes at the centre
#fig = plt.figure()
# ax = fig.add_subplot()
# ax.spines['left'].set_position('center')
# ax.spines['bottom'].set_position('center')
# ax.spines['right'].set_color('none')
# ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.yaxis.set_ticks_position('left')
# plot the function
plt.subplot(2,2,2)
plt.plot(x2,y2, 'g', 
         marker = 'o',
        label = '$x^{3}$')
plt.title('$x^{3}$')

plt.legend()


plt.subplot(2,2,3)
plt.plot(x3,y3,'r', 
         marker = 'v',
        label = '$2^{x}$')
plt.title('$2^{x}$')

plt.legend()

plt.subplot(2,2,4)
plt.plot(x4,y4, 'c', 
         marker = 's',
        label = '${1}/{(x+1)}$')
plt.title('${1}/{(x+1)}$')

plt.suptitle('Various Math Functions', fontsize = 20)
plt.tight_layout()
plt.legend()

;
```
